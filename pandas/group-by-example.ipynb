```python
import pandas as pd


ipl_data = {'Team': ['Riders', 'Riders', 'Devils', 'Devils', 'Kings',
   'kings', 'Kings', 'Kings', 'Riders', 'Royals', 'Royals', 'Riders'],
   'Rank': [1, 2, 2, 3, 3,4 ,1 ,1,2 , 4,1,2],
   'Year': [2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
   'Points':[876,789,863,673,741,812,756,788,694,701,804,690]}
df = pd.DataFrame(ipl_data)

print (df)
```

```python
grouped = df.groupby('Year')
print (grouped)

#Nature of groups
for name,group in grouped:
   print (name)
   print (group)
```

```python
import numpy as np
#Column selection
subgroup = df.groupby('Year')["Rank"]
print (subgroup)
type(subgroup)

#By chosing a column, no aggregation is not done yet
#It is merely positiong the column in the "tree"
#it is still a tree structure
#Its type accordingly is "SeriesGroupBy"
#it is not series yet (so to say, what ever the nature of a series is)
#That is for another day
aggregatedSubgroup = df.groupby('Year')["Rank"].agg(np.mean)
print(aggregatedSubgroup)
print(type(aggregatedSubgroup))

#Covert the series to a dataframe
subdf = aggregatedSubgroup.to_frame()
print(subdf)
```

```python
aggregatedSubgroup2 = df.groupby('Year')[["Rank","Points"]].agg(np.mean)
print(aggregatedSubgroup2)
print(type(aggregatedSubgroup2))

#aggregatedSubgroup2.describe()
aggregatedSubgroup2.head()
```
