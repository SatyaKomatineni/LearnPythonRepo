# Imports

```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd

#A list of days and temperatures
daysList = [1,2,3,4,5,6,7]
TempList = [50,51,52,48,47,49,46]

#A hashtable<colname, column-value-list>
data = {"Day": daysList, "Temperature": TempList}

#init the dataframe
df = pd.DataFrame(data)

#Print the dataframe
df
```

```python
plt.figure(figsize=[6,3])
plt.xlabel("Day")
plt.ylabel("Temp")
x = plt.plot(df["Day"],df["Temperature"],linewidth=4)
print(x)
```
